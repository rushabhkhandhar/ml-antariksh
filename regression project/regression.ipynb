Importing pandas 

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

Reading CSV file

```python
CO2Emission=pd.read_csv('CO2 Emissions_Canada.csv')
```

Droping a model column

```python
CO2Emission.drop(columns=['Model','Make','Vehicle Class','Transmission'],inplace=True)
```

```python
CO2Emission
```

```python
CO2Emission=pd.get_dummies(CO2Emission,columns=['Fuel Type'],dtype=int)
CO2Emission
```

test-split

```python
CO2Emission_X=(CO2Emission.drop('CO2 Emissions(g/km)',axis=1))
CO2Emission_Y=(CO2Emission['CO2 Emissions(g/km)'])
```

```python
CO2Emission_train,CO2Emission_test,CO2Emission_Y_train,CO2Emission_Y_test=train_test_split(CO2Emission_X,CO2Emission_Y,random_state=10,test_size=0.2)
```

```python
print(CO2Emission_train)
print(CO2Emission_test)
```

```python
model=LinearRegression()
model.fit(CO2Emission_train,CO2Emission_Y_train)
```

```python
mean_squared_error(CO2Emission_Y_train, model.predict(CO2Emission_train))
```

```python
mean_squared_error(CO2Emission_Y_test, model.predict(CO2Emission_test))
```
